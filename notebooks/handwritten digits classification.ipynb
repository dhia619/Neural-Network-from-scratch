```python
import pandas as pd
import numpy as np
from network import Network
```

## Data preparation ##

```python
from sklearn.datasets import fetch_openml

# Fetch the MNIST dataset
mnist = fetch_openml('mnist_784', version=1)
x, y = mnist["data"], mnist["target"]
```

```python
x =  x / 255
y = y.astype("int64")
```

```python
x.head(10)
```

```python
data = pd.concat([x,y],axis=1)
```

```python
data.head()
```

```python
randomized_data = data.sample(frac=1,random_state=42)

split_index = int(0.8 * len(randomized_data))

train_data = randomized_data[:split_index]
test_data = randomized_data[split_index:]

train_x,train_y = train_data[["pixel"+str(i)for i in range(1,785)]],train_data["class"]
test_x,test_y = test_data[["pixel"+str(i)for i in range(1,785)]],test_data["class"]
```

```python
#one hot encode the labels
train_y = pd.get_dummies(train_y)
train_y = train_y.astype("int")
train_x,train_y = train_x.to_numpy(),train_y.to_numpy()
test_y = pd.get_dummies(test_y)
test_y = test_y.astype("int")
test_x,test_y = test_x.to_numpy(),test_y.to_numpy()
```

## Experimenting ##

```python
#create an instance of the network
model = Network([784,128,128,10])
```

```python
model.train(train_x.T,train_y.T,learnRate=0.01,epochs=30,batch_size=100,show_training_progress=True,show_training_progress_rate=5)
```

```python
print(model.calculate_accuracy(train_y.T,model.feed_forward(train_x.T))*100,"%")
```

```python
print(model.calculate_accuracy(test_y.T,model.feed_forward(test_x.T))*100,"%")
```

```python
model.savemodel("handwritten_digits_V1.0")
```
